Principal Component Analysis (PCA) ek dimensionality reduction technique hai jo data ke features (columns) ki size kam karne ke liye use hoti hai, bina jyada information lose kiye. Ye mainly high-dimensional data ko simplify karne aur fast analysis ke liye use hoti hai.

PCA kaise kaam karta hai?

Standardization – Sabhi features ko normalize kiya jata hai taki unka mean 0 aur variance 1 ho jaye.
Covariance Matrix – Features ke beech ka relation check karne ke liye covariance matrix calculate kiya jata hai.
Eigenvalues & Eigenvectors – Covariance matrix se eigenvalues aur eigenvectors nikale jate hain jo important directions (principal components) dikhate hain.
Sorting Principal Components – Eigenvalues ko descending order me arrange kiya jata hai, jisme highest variance wale components ko rakha jata hai.
Feature Selection – Top k components choose kiye jate hain jo maximum information retain karein.
Data Transformation – Original data ko principal components ke basis par project kiya jata hai taki dimension kam ho jaye.

PCA kyun use karte hain?
✅ Dimensionality reduce karta hai, jisse visualization aur computation fast ho jati hai.
✅ Multicollinearity hata deta hai, matlab correlated features ko independent banata hai.
✅ Machine learning models fast aur optimized ho jate hain.
✅ Overfitting reduce karta hai, kyunki extra features remove ho jate hain.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
data = pd.DataFrame(digits.data,columns=digits.feature_names)
data
```

```python
data['target'] = digits.target
```

```python
data
```

```python
X = data.drop(['target'],axis='columns')
X
```

```python
y = data.target
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
X = pd.DataFrame(X,columns=digits.feature_names)
X    
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model =  LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)*100
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95)
```

```python
x_pca = pca.fit_transform(X)
x_pca = pd.DataFrame(x_pca)
x_pca
```

```python
pca.explained_variance_ratio_
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x_pca,y,test_size= 0.2)
```

```python
model.fit(X_train,y_train)
model.score(X_test,y_test)*100
```

```python
pca.n_components_
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(X)
x_pca = pd.DataFrame(x_pca)
x_pca
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x_pca,y,test_size= 0.2)
```

```python
model.fit(X_train,y_train)
model.score(X_test,y_test)*100
```
